### BMR Ticket Price Prediction

Big Mountain Resort (hereafter, BMR) requested that the Data team identify optimal pricing for adult ticket prices, given the increase expenditures in the prior season. In particular, BMR increased the operation costs by 1.5 million due to the addition of a chair lift. Currently, BMR charges `$81` per ticket; our modeled price is `$90.90` (an increase of `$9.90`). This would result in a an expected annual revenue of 17,322,717 if the resort attracts the approximated 350,000 visitors in the coming year. 


**The following four adjustments are recommended to justify the change in price:**

1. add an additional run
2. increase the vertical drop of the longest run by 150 ft
3. guarantee 2 additional acres of snow making ability
4. add an additional fast quad chair lift

#### Random Forest | Feature Relevance

When preparing our model to predict a price increase, we used Random Forest Regression to identify attributes relevant to ticket price. The model returned vertical drop, snow making,  runs, and fast quads as the most important features:



#### Potential to Reduce Costs: Closing Runs 

<span style = "color:indianred"> **This option is not recommended due to unknown potential loss of visitors, as well as reduced support for ticket prices.** </span>

BMR could close up to 5 runs, thus reducing operating costs. However, the model indicates that this would reduce support for ticket price by `$1.25` per ticket (by closing a maximum of 5 runs). This would result in a net loss of revenue of `$437,500`. We would need to further explore the potential loss of visitors were 5 runs to be closed as well as the amount of savings in operational costs.



#### Increased Price Support: Adding Features

<span style = "color:mediumseagreen"> **This option is recommended due to increased support for ticket prices, as well as potential to attract more visitors.** </span>

Our regression model was run by adding adding one run, 150ft in max vertical drop, one fast-quad chair lift, and 2 acres of snow making ability. 

This scenario increases support for ticket price by `$9.90` <br>
Over the season, this could be expected to amount to `$17,322,717`

> To confirm the strategy, this was compared with increasing the longest run by .2 miles and adding 4 acres of snow making ability, thus leaning into guaranteed snow alone. Interestingly, this made no difference in support for ticket prices. Thus, the above strategy is recommended.
